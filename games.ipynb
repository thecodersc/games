```python
import pandas as pd
```

```python
file = '/home/stanislaw/Documents/work/Best_selling_games_in_history.csv'
```

```python
dane = pd.read_csv(file)
```

```python
dane.head(5)
```

```python
dane = dane.drop('Ref.',axis=1)
dane.head()
```

```python
dane.groupby(by='Publisher').agg('sum').sort_values(by='Sales',ascending=False)
```

```python
dane.groupby(by='Platform').agg('sum').sort_values(by='Sales',ascending=False)
```

```python
dane.dtypes
```

```python
dane['Release date'] = pd.to_datetime(dane['Release date'])
```

```python
dane['year'] = dane['Release date'].map(lambda x: x.strftime('%Y'))
```

```python
dane
```

```python
dane.groupby(by='year').agg('sum').sort_values(by='Sales',ascending=False)
```
